```python
import pandas as pd

# Membaca file excel pada sheet_name pemakaian_mesin
df = pd.read_excel("soal.xlsx", "pemakaian_mesin")
df
```

```python
df.info()
```

```python
# Menambahkan kolom Durasi kerja mesin yang menunjukkan lama waktu mesin bekerja
df['Durasi Kerja Mesin'] = df['Selesai'] - df['Mulai']
df
```

```python
df.info()
```

```python
from datetime import timedelta
```

```python
# Memeriksa apakah ada mesin yang bekerja lebih dari 1 hari
df[df['Selesai']-df['Mulai'] > timedelta(days=1)]
```

Dari tabel di atas diperoleh informasi bahwa terdapat mesin yang bekerja lebih dari satu hari dan semuanya sudah pasti masuk ke shift 1 pada tanggal mesin mulai bekerja dan shift 1 pada tanggal selanjutnya. Karena terdapat mesin yang bekerja lebih dari satu hari, maka dalam pembagian shift harus diperiksa apakah mesin tersebut bekerja lebih dari satu hari. Jika mesin bekerja lebih dari satu hari, maka mesin tersebut akan masuk ke shift 1 di hari selanjutnya. 

```python
tanggal_shift = []
for i in range(len(df)):
    # Mendapatkan waktu mulai kerja mesin
    waktu_mulai = df.loc[i]['Mulai'] - df.loc[i]['Tanggal Mulai']
    # Memeriksa apakah mesin mulai bekerja pada shift 1
    total_durasi = timedelta(0)
    if (waktu_mulai >= timedelta(hours=6) and waktu_mulai < timedelta(hours=14)):
        durasi = timedelta(hours=14) - waktu_mulai if df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(hours=14) else (df.loc[i]['Durasi Kerja Mesin'])
        total_durasi += durasi
        tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), 1, durasi])
        
        # Pengecekan apakah masuk shift 2
        if (df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(hours=14)):
            durasi = timedelta(hours=8) if df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(hours=8) else (df.loc[i]['Durasi Kerja Mesin'] - total_durasi)
            total_durasi += durasi
            tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), 2, durasi])

        # Pengecekan apakah masuk shift 3
        if (df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(hours=22)):
            durasi = timedelta(hours=8) if df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(hours=8) else (df.loc[i]['Durasi Kerja Mesin'] - total_durasi)
            total_durasi += durasi
            tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), 3, durasi])
        
        # Pengecekan apakah masuk ke shift 1 di hari selanjutnya
        if (df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(days=1, hours=6)):
            tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Selesai'].date(), 1, df.loc[i]['Durasi Kerja Mesin'] - total_durasi])
        
    # Memeriksa apakah mesin mulai bekerja pada shift 2 
    elif (waktu_mulai >= timedelta(hours=14) and waktu_mulai < timedelta(hours=22)):
        durasi = timedelta(hours=22) - waktu_mulai if df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(hours=22) else (df.loc[i]['Durasi Kerja Mesin'])
        total_durasi += durasi
        tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), 2, durasi])
        # Pengecekan apakah masuk shift 3
        if (df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(hours=22)):
            durasi = timedelta(hours=8) if df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(hours=8) else df.loc[i]['Durasi Kerja Mesin'] - total_durasi
            total_durasi += durasi
            tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), 3, durasi])
            if (durasi > timedelta(hours=8)):
                print(durasi, '1')
        if (df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(days=1, hours=6)):
            tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Selesai'].date(), 1, df.loc[i]['Durasi Kerja Mesin'] - total_durasi])

    else: # Jika waktu mesin mulai bekerja bukan pada shift 1 dan 2, maka pasti masuk shift 3
        if (waktu_mulai >= timedelta(hours=22) and waktu_mulai <= timedelta(hours=23, minutes=59, seconds=59)):
            durasi = timedelta(days=1, hours=6) - waktu_mulai if df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(days=1, hours=6) else (df.loc[i]['Durasi Kerja Mesin'])
        else:
            durasi = timedelta(hours=6) - waktu_mulai if df.loc[i]['Durasi Kerja Mesin'] + waktu_mulai > timedelta(hours=6) else (df.loc[i]['Durasi Kerja Mesin'])

        total_durasi += durasi
        tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), 3, durasi])
        
        shift = 1
        if (df['Tanggal Mulai'][i] == df['Tanggal Selesai'][i]):
            while (df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(hours=8)):
                tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), shift, 8])
                shift += 1
                total_durasi += timedelta(hours=8)
            if (df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(0)):
                tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), shift, df.loc[i]['Durasi Kerja Mesin'] - total_durasi])
        else:
            while (df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(hours=8)):
                tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Mulai'].date(), shift, timedelta(hours=8)])
                shift += 1
                total_durasi += timedelta(hours=8)
            if (df.loc[i]['Durasi Kerja Mesin'] - total_durasi > timedelta(0)):
                tanggal_shift.append([df.loc[i]['Mesin'], df.loc[i]['Ruangan'], df.loc[i]['Tanggal Selesai'].date(), 1, df.loc[i]['Durasi Kerja Mesin'] - total_durasi])
```

```python
# Melihat panjang data
len(tanggal_shift)
```

```python
# Membuat dataframe dari data tanggal_shift
df_tanggal_shift = pd.DataFrame(data=tanggal_shift, columns=['Mesin', 'Ruangan', 'Tanggal', 'Shift', 'Durasi'])
df_tanggal_shift
```

```python
# Mengubah format durasi menjadi integer yang menunjukkan lamanya waktu operasi mesin dalam menit pada shift tersebut
df_tanggal_shift['Durasi (menit)'] = df_tanggal_shift['Durasi'].apply(lambda x: round(float(x.seconds) / 60.0))
df_tanggal_shift.drop(['Durasi'], axis=1)
df_tanggal_shift
```

```python
# Save dataframe ke excel
df_tanggal_shift.to_excel("Shift.xlsx", index=False)
```
